```python
df1= spark.read.option('header','True').csv('bill_mst.csv',inferSchema=True)
```

```python
df1.show(df1.count(),False)
```

#Find out High value bills having bill amount equal to or more than Rs. 180000. Save the result as HighValueBills.


```python
HighValueBills=df1.filter(df1['BILL_AMT']>180000)
```

```python
HighValueBills.show()
```

#Find out Instances of bill amount and approved amount being different. Save the result as BillPassDiff

```python
BillPassDiff=df1.filter(df1['BILL_AMT']!=df1['PASS_AMT'])
```

```python
BillPassDiff.show()
```

#Bills which have been booked in during the month of March 2011. Save the result as BillsBookedMar11.

```python
from pyspark.sql.functions import to_date, lit, col

# Assuming your BILL_DATE is in "dd/MM/yyyy" format
df1 = df1.withColumn("BILL_DATE", to_date(col("BILL_DATE"), "dd/MM/yyyy"))

# Filter rows where BILL_DATE is in March 2011
BillsBookedMar11 = df1.filter((df1["BILL_DATE"] >= lit("2011-03-01")) & (df1["BILL_DATE"] <= lit("2011-03-31")))
```

```python
BillsBookedMar11.show()
```
